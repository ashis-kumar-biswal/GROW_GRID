1. Write a python program using inbuilt function to Remove all consonants from each string

```python
s = ["computer", "science", "python", "programming"]
vowels = "aeiouAEIOU"
result = ""
for i in s:
    for ch in i:
        if ch in vowels:
            result += ch
print(result)
```

2. Write a python program using inbuilt function to Filter strings that alternate vowels and consonants

```python
# Write a python program using inbuilt function to Filter strings that alternate vowels and consonants
words = ["baba", "mama", "code", "idea", "apple", "banana"]
vowels = "aeiouAEIOU"
def alterword(word):
    for i in range(len(word) - 1):
        if (word[i] in vowels) == (word[i + 1] in vowels):
            return False
    return True
result = list(filter(lambda x: not alterword(x), words))
print(result)
```

3. Write a python program using inbuilt function to Keep strings that contain at least 3 different vowels

```python
words = ["Ashis", "Apple", "orange", "Manish", "mountaineer", "poem", "Kai"]
vowels = "aeiouAEIOU"
def check(word):
    count = 0
    for v in vowels:
        if word.count(v) > 0:
            count += 1
    return count >= 3
result = list(filter(check, words))
print(result)
```

4. Write a python program using inbuilt function to Build a string of the first character from each word

```python
words = ["Mountain", "Ashis", "Manish", "Balloon", "Apple"]
result = "".join(map(lambda word: word[0], words))
print(result)
```

5. Write a python program using inbuilt function to Reduce list to a string of only digits extracted from all words

```python
words = ["apple1", "2banana3", "cherry", "45date"]
s1 = "".join(words)
s2 = filter(str.isdigit, s1)
s3 = "".join(s2)
print("Extracted digits:", s3)
```

6. Write a python program using inbuilt function to Double the word at every index divisible by 4

```python
def double_index4(x):
    return list(map(lambda x: x[1]*2 if x[0] % 4 == 0 else x[1], enumerate(x)))
words = ["apple", "ball", "cat", "dog", "eye", "frog", "ronaldo", "hat", "ipad"]
print(double_index4(words))
```

7.From a sentence, return all words that appear more than once

```python
sentence = "red blue green red yellow blue white red"
words = sentence.split()
result = list(filter(lambda x: words.count(x) > 1, set(words)))
print(result)
```

8. Write a python program using inbuilt function to Return list of (index, word) for words that are palindromes

```python
def palindrome(x):
    return list(filter(lambda x: x[1] == x[1][::-1], enumerate(x)))
words = ["madam", "ashis", "level", "cuttack"]
print(palindrome(words))
```

9. Write a python program using inbuilt function to Sort by proximity to a target length (e.g., 5)

```python
words = ["ashis", "apple", "orange", "me", "artic"]
result = sorted(words, key=lambda x: abs(len(x) - 5))
print(result)
```

10. Write a python program using inbuilt function to Filter strings that match their sorted version (alphabetical)

```python
words = ["aap", "bjp", "tmc", "abc", "xyz"]
result = list(filter(lambda x: x != "".join(sorted(x)),words))
print(result)
```
