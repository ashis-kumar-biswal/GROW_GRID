```python
def c_vowel(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
text = input("Enter a string: ")
print("Number of vowels:", c_vowel(text))
```

```python
#Take a list of numbers positive, neagative, zero all included and filter out the positive numbers from the list.
def positive(num):
    return num > 0
numbers = [10, -5, 0, 23, -8, 15, -2, 7, 0, -1]
result = filter(positive, numbers)
print(list(result))
```

```python
#Take a tuple of numbers and multiply each number in the tuple with 5
def multiply(num):
    return num * 5
numbers = (10, 20, 30, 40, 50)
result = map(multiply, numbers)
print(list(result))
```

```python
#Find the maximum number in a list using reduce
from functools import reduce
def maximum(x, y):
    if x > y:
        return x
    return y
numbers = [10, -5, 0, 23, -8, 15, -2, 7, 0, -1]
result = reduce(maximum, numbers)
print(result)
```

```python
# Take a list of numbers and add 3 to the numbers divisible by 3 in that list.
def add_three(num):
    if num % 3 == 0:
        return num + 3
    return num
numbers = [3, 5, 6, 8, 9, 12, 15, 20]
result = map(add_three, numbers)
print(list(result))
```

```python
#Take a list of numbers, even and odd both, and then find the product of only even numbers
from functools import reduce
def even(num):
    return num % 2 == 0
numbers = [2, 3, 4, 5, 6, 7, 8, 9]
even_numbers = filter(even, numbers)
def product(x, y):
    return x * y
result = reduce(product, even_numbers)
print(result)
```
