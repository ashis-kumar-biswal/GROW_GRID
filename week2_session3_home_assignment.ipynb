1. Write a python function to Capitalize all words in a list of strings

```python
words = ["cregen", "rob", "ned"]
result = list(map(lambda x: x.capitalize(), words))
print(result)
```

2. Write a python function to Convert list of strings to their lengths

```python
words = ["stalin", "lenin", "putin"]
result = list(map(len, words))
print(result)
```

3. Write a python function to Filter out strings that are palindromes

```python
words = ["madam", "ashis", "aba", "cat"]
result = list(filter(lambda x: x != x[::-1], words))
print(result)
```

4. Write a python function to Concatenate all strings into one

```python
from functools import reduce
words = ["Vallar", "Morghulis"]
result = reduce(lambda x, y: x + y, words)
print(result)
```

5. Write a python function to Find the longest string in a list

```python
words = ["Starks", "Lanisters", "Baratheons", "targaryens"]
result = max(words, key=len)
print(result)
```

6. Write a python function to Create a list of tuples (index, string) for strings starting with 'b'

```python
words = ["apple", "all", "batman", "baratheons", "banana", "armour"]
result = list(filter(lambda x: x[1].startswith("b"), enumerate(words)))
print(result)
```

7. Write a python function to Find index of the longest string

```python
words = ["aegon", "maeker", "loyonel", "deamon"]
result = max(enumerate(words), key=lambda x: len(x[1]))
print(result)
```

8. Write a python function to Reverse each string in the list

```python
words = ["daemon", "catelyn", "john"]
result = list(map(lambda x: x[::-1], words))
print(result)
```

9. Write a python function to Get ASCII values of characters in a string

```python
text = "NightKing"
result = list(map(ord, text))
print(result)
```

10. Write a python function to Filter strings that start with a vowel

```python
words = ["ramen", "sushi", "blood", "umbrella", "cat"]
result = list(filter(lambda x: x[0].lower() not in "aeiou", words))
print(result)
```
